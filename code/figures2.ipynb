```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import scipy
import numpy as np

from basic import generate_erdos_renyi_graph, perturb_graph
from algos import spectral_method, variance_method
```

```python
p = 0.3
q = 0.6
epsilon = 0.1
trials = 20

var_errors_detailed = []
var_errors = []
ns = np.arange(150, 1501, 150)
for n in ns:
    var_error = 0
    for _ in range(trials):
        G, _ = generate_erdos_renyi_graph(n, p)
        p_original = nx.average_clustering(G)

        preturbed_G = perturb_graph(G, epsilon, q)
        
        p_hat, var = variance_method(preturbed_G, epsilon)
        error = abs(p_hat - p_original)
        var_error += error
        var_errors_detailed.append((n, error))
    var_errors.append(var_error / trials)
    print(f"n: {n}\nerror:\t{var_error / trials}\n")


plt.plot(ns, var_errors)
plt.xlabel("Number of Nodes")
plt.ylabel("Error")
plt.title("Variance Method: Error vs. Number of Nodes")
plt.show()

print(var_errors)
print(var_errors_detailed)
```
